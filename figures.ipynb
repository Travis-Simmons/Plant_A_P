```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# root shoot biomase ratio ANOVA

cw_lm=ols('root_shoot_ratio ~ C(treatment)', data=final_day).fit() #Specify C for Categorical
print(sm.stats.anova_lm(cw_lm, typ=2))
```

```python
# plant moisture content ANOVA

cw_lm=ols('plant_moisture_content ~ C(treatment)', data=final_day).fit() #Specify C for Categorical
print(sm.stats.anova_lm(cw_lm, typ=2))
```
