```python
import pandas as pd
import seaborn as sns
```

```python
final_day = pd.read_csv('final_day.csv')
leaf_counts = pd.read_csv('leaf_counts.csv')
```

```python
final_day.head()
```

```python
leaf_counts.head()
```

```python
leaf_counts['treatment'] = leaf_counts['pod'].str[:-1]
leaf_counts['pod_number'] = leaf_counts['pod'].str[-1:]
```

```python
final_day['pod_number'] = final_day['pod'];final_day
final_day['pod'] = final_day['treatment'] + final_day['pod'].astype(str)
final_day
```

```python
leaf_counts
```

```python


final_day_leaf_count  = final_day.groupby('pod').sum()
```

```python
final_day_leaf_count.reset_index(inplace = True)
```

```python
leaf_counts['10/28'] = final_day_leaf_count['leaf_pair']
```

```python
leaf_counts
```

```python
# Data ready to use

final_day.head()
```

```python
leaf_counts.head()
```

```python
leaf_melt = leaf_counts.melt(id_vars= ['treatment', 'pod_number', 'pod']);leaf_melt
leaf_melt['variable'] = leaf_melt['variable']+'/21';leaf_melt
leaf_melt['variable'] = pd.to_datetime(leaf_melt['variable'], infer_datetime_format=True);leaf_melt
leaf_melt  = leaf_melt.groupby(['treatment', 'variable']).mean();leaf_melt



```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (15,8))
ax = sns.lineplot(data=leaf_melt, x="variable", y="value", hue="treatment")
plt.xticks(rotation = 45)
plt.xlabel('Date')
plt.ylabel('Average Number of Leaf Pairs Present')
# plt.title('Average Number of Leaf Pairs by Treatment Over Time', fontsize = 16)
```

```python
grouped_final = final_day.groupby('treatment').mean();grouped_final
grouped_final.reset_index(inplace = True);grouped_final
```

```python
# plt.figure(figsize = (15,8))
import numpy as np
ax = sns.barplot(data=grouped_final, x="treatment", y="root_length" )
plt.xticks(rotation = 45)
plt.xlabel('Treatment')
plt.ylabel('Average Root Length')
# plt.title('Average Root Length by Treatment', fontsize = 16)
```

```python
test_final = grouped_final[['treatment', 'root_number', 'node']]
```

```python


plt.figure(figsize = (15,8))
ax = test_final.plot(kind = 'bar', x = 'treatment')
plt.xticks(rotation = 1)
plt.xlabel('Treatment')
plt.ylabel('Average Node Number vs Root Number')
# plt.title('move to bottom', fontsize = 16)
```

```python
grouped_final['root_shoot_ratio'] = grouped_final['root_length']/ grouped_final['shoot_length'];grouped_final
```

```python
plt.figure(figsize = (15,8))
ax = grouped_final.plot(kind = 'bar', x = 'treatment', y = 'root_shoot_ratio', cmap = 'tab10' , legend = False)
plt.xticks(rotation = 1)
plt.xlabel('Treatment')
plt.ylabel('Root Shoot Ratio')
# plt.title('move to bottom', fontsize = 16)
```
